EX 1:Implementati algoritmul **Metodei Puterii** si testati pe o matrice de dimensiune ```n=6```.
Verificati rezultatul folosind ```numpy.linalg.eig()```



````python
import numpy as np

def metoda_puterii(matrix, iteratii=1000):
  n = len(matrix)
  y = np.random.rand(n)

  for _ in range(iteratii):
    w = np.dot(matrix, y)
    norm = np.linalg.norm(w, 2)
    y = w / norm
    val_proprie = np.dot(np.dot(y, matrix), y)

    return val_proprie

# TestÄƒm pe o matrice de dimensiune 6x6
n = 6
A = np.random.rand(n, n)
rezultat_MP=metoda_puterii(A)

eigenvalues_numpy, eigenvectors_numpy= np.linalg.eig(A)
max_eigenvalue_numpy = max(eigenvalues_numpy)

print("Rezultatul Metodei Puterii:")
print("Valoarea proprie:", rezultat_MP)
print("\nRezultatul folosind numpy.linalg.eig():")
print("Valoare proprie:", max_eigenvalue_numpy)
````

EX 2:
Fisierul `grafuri.pickle` contine matricile de adiacenta a 3 grafuri.

1.   Obtineti grafurile asociate matricilor si vizualizati grafurile
2.   Calculati valorile proprii
3.   Verificati daca grafurile sunt complete, daca sunt bipartite si care e dimensiunea celei mai mari clici



````python
import pickle
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
````

````python
with open('grafuri.pickle', 'rb') as f:
    matrices = pickle.load(f)
    print(matrices)
````

````python
# Afisati matricile
for matrix in matrices:
  #print(f"Matrix {i + 1}:\n{matrix}\n")
  graf = nx.Graph(matrix)
  nx.draw(graf, with_labels=True)
  plt.show()
````

````python
val_proprii=[]
for i,matrix in enumerate(matrices):
  value=np.linalg.eigvals(matrix)
  val_proprii.append(value)

for i,val in enumerate(val_proprii):
  print(val)
````

````python
for i,matrix in enumerate(matrices):
  graph = nx.Graph(matrix)
  is_complete = np.all(np.logical_or(matrix == 1, np.eye(matrix.shape[0], dtype=int)))
  print(f"Graf {i + 1} este complet: {is_complete}")

  is_bipartite = nx.is_bipartite(graph)
  print(f"Graf {i + 1} este bipartit: {is_bipartite}")

  clici = list(nx.find_cliques(graph))
  max_clic_size = max(len(clic) for clic in clici)
  print(f"Dimensiunea celei mai mari clici pentru Graf {i + 1}: {max_clic_size}\n")
````
