```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
```

Creati o matrice aleatoare A cu 10 randuri, 4 coloane. Calculati rangul matricii

```python
m = 10
r = 4
A = np.random.rand(m, r)
print(A)
rank_A = np.linalg.matrix_rank(A)

print("\nRangul matricei A:", rank_A)
```

Adauga 4 coloane linear dependente. Pentru acrea noile coloane, luati o combinatie liniara de c=3 coloane deja
existente in A

```python
c = 3
coloane = np.sum(A[:, :c], axis=1, keepdims=True) * np.random.rand(1, 4)
A_extinsa= np.hstack((A, coloane))

rank_A_extinsa = np.linalg.matrix_rank(A_extinsa)

print("Matricea A extinsa:")
print(A_extinsa)
print("\nRangul matricei A extinsa:", rank_A_extinsa)
```

Adaugati la matricea obtinuta anterior un zgomot Gaussian de medie 0 si dispersie 0.2
Calculati rangul noii matrici

```python
noise = np.random.normal(0, 0.2, A_extinsa.shape)
A_zgomot = A_extinsa + noise
rank_A_zgomot=np.linalg.matrix_rank(A_zgomot)
print("Matricea A cu zgomot:")
print(A_zgomot)
print("Rang A_zgomot:", rank_A_zgomot)
```

Calculati descompunerea valorilor singulare pentru matricea de mai sus si afisati valorile singulare. Care din aceste valori sunt neglijabile? Puteti deduce rangul matricii neafectate de zgomot din aceste valori?

```python
U, val_singulare, Vt = np.linalg.svd(A_zgomot)
print("SVD de A_zgomot:")
print("U:")
print(U)
print("\nval_singulare:")
print(np.diag(val_singulare))
print("\nVt:")
print(Vt)
```

```python
# Setam o valoare prag pentru a considera valorile singulare ca fiind neglijabile
prag = 1e-1

neglijabile = val_singulare < prag

# Numarul de valori singulare neglijabile
numar_neglijabile = np.sum(neglijabile)

print(f"\nNumarul de valori singulare neglijabile: {numar_neglijabile}")

# Deduce rangul matricii neafectate de zgomot
rang_neafectat_de_zgomot = len(val_singulare) - numar_neglijabile

print(f"\nRangul matricii neafectate de zgomot: {rang_neafectat_de_zgomot}?")
```

```python
image=plt.imread("words2023.png")
```

```python
plt.imshow(image)
```

```python
if image.ndim == 3:
    # Convertim imaginea color n imagine alb-negru
    image = np.mean(image, axis=-1)
```

```python
U, val_singulare, Vt = np.linalg.svd(image, full_matrices=False)
```

```python
plt.subplot(122)
plt.imshow(image, cmap='gray')
plt.title('Imaginea incarcata')

# Afim primele 10 valori singulare
plt.subplot(121)
plt.plot(val_singulare[:10])
plt.title('Primele 10 Valori Singulare')
plt.xlabel('Numarul de valori singulare')
plt.ylabel('Valoarea singulara')

plt.show()
```

```python
# Afim imaginea original i aproximarea de rang k
plt.imshow(image, cmap='gray')
plt.title('Imaginea original')

```

```python
k=50
# Trunchiati valorile singulare i recompuneti matricea aproximativa
approx_matrix = np.dot(U[:, :k], np.dot(np.diag(val_singulare[:k]), Vt[:k, :]))

plt.imshow(approx_matrix, cmap='gray')
plt.title(f'Aproximare de rang {k}')

plt.show()
```
